El clculo de la potencia elica es de vital importancia en la evaluacin del recurso elico.

La energa cintica del viento puede calcularse si se supone que el aire pasa por un cilindro , que encierra un rea A a una velocidad de viento V.

$$\rightarrow P=\frac{E_c}{s}=\frac{1}{2}\rho AV^3$$

La densidad de potencia se define como la potencia por unidad de rea. $\frac{P}{A}=[W/m^2]$

Y como es de esperar existen diversas formas de calcular la el recurso elico, esto siempre depende de la presicin y la cantidad de nuestros datos. 

### Uso directo de datos
 La densidad de potencia promedio por unidad de rea la calculamos como:
 
 $\overline{P}/A=\frac{1}{2}\rho \frac{1}{N} \sum_{i=1}^N AU_i^3$

Ahora la densidad de energa por unidad de rea en un tiempo $N\Delta t$.

 $\overline{E}/A=\frac{1}{2}\rho \frac{1}{N} \sum_{i=1}^N AU_i^3 N\Delta t=\frac{\overline{P}}{A} N\Delta t$ 


La potencia media del aerogenerador $\overline{P_w}$ es:

$\overline{P_w}=\frac{1}{N} \sum_{i=1}^N P_w(U_i)$

Con $P_w$ es la potencia de salida definida por la curva de potencia del aerogenerador.

Por lo tanto la energa del aergenerador, $E_w$:

$$\Rightarrow E_w= \sum_{i=1}^N P_w(U_i) \Delta t$$

### Tcnicas estdisticas

Para esto antes habr que elegir una buena funcin de probabilidad $p(U)$ que se ajuste a los datos de viento de la localidad y si se cuenta con la curva de potencia del arogenerador de interes entonces, $P_w$, entonces la potencia la potencia media,$\overline{P_w}$, del arogenerador ser:

$\overline{P_w}= \int_{0}^\infty P_w(U)p(U)dU$


Al histograma de las velocidades del viento se le debe ajustar una distribucin que modele nuestros datos, para ello es importante el hecho que usemos la mejor distribucin. Usar una distribucin Weibul ampliamente utilizada en el campo, ya que una de sus ventajas es el hecho que inicia en cero, pensando a las "velocidades del viento como positivas" es decir, solamente sus magnitudes, datos que tenemos.


Una variable x tiene una distribucin Weibul si su funcin de probabilidad f tiene la forma:

$$f(x)=\frac{\alpha}{\beta}(\frac{X}{\beta})^{\alpha-1} e^{-(\frac{X}{\beta})^{\alpha}}$$ con $$\alpha,\beta,X>0$$

Con $X$ la velocidad promedio,

$\alpha$ un parmetro de forma,

y $\beta$ un parmetro de escala.

```python
#Fit weibul normal distribution
par= weibull_min.fit(df.WS_80mA_mean)

# Plot Histogram
plt.hist(df.WS_80mA_mean, density=True)

# Plot weibull
xmin, xmax = plt.xlim()
x = np.linspace(0, xmax, 100)
p = weibull_min.pdf(x, par[0], par[1],par[2])
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: shape = %.2f,  loc = %.2f, scale = %.2f" % (par[0], par[1],par[2])
plt.title(title)

plt.show()
```
